In my marketing class, we have spoken in depth about brand naming, brand positioning, marketing, target segements, and everything in-between. It made me think about how good an LLM would be at doing this. Could an LLM ensure no terrible brand decisions were made? Could they be a company's CMO? 

On a smaller scale - for the sake of this blogpost - I wanted to see if an LLM could do what creative agencies spend weeks doing: generate names that sound human, memorable, and just a little bit clever. 

---

I gave ChatGPT a simple brief:

“You are a brand strategist. Generate 10 names for a luggage company that’s stylish, functional, and appeals to young travelers.”

The results came back instantly:

Glide

JetSet

Nomadica

CarryOn & Co.

The Atlas Project

Not bad, but also… not good. They sounded like ideas from a brainstorming session that ended very late. The model clearly understood form — short, sleek, aspirational — but not feeling.

---

Next, I added constraints:

“The brand should feel luxurious but minimal. Avoid clichés like ‘jet,’ ‘nomad,’ or ‘travel.’ Think Béis meets Apple.”

Suddenly, the tone shifted:

Paravel

Onyx

Vale

In Transit

Haul

Still imperfect, but closer. It started understanding what I meant by vibe — the emotional texture behind a brand, not just the dictionary meaning. When I added follow-up instructions like “make it sound feminine but not fragile” or “use soft phonetics,” it began producing names that actually sounded on-brand.

---

The real magic appeared when I turned the prompt into a conversation:

Me: “That sounds a bit too techy.”
LLM: “Would you prefer something softer, maybe inspired by natural materials?”
Me: “Yes — something that feels tactile, not digital.”
LLM: “How about Loom, Drift, or Sonder?”

At that point, it didn’t feel like a generator. It felt like a collaborator.

<p style="color:red">What I Learned</p>

LLMs aren’t great at originality on their own — but they’re excellent mirrors for refining creative direction. They give you quick volume, variety, and vocabulary, which you can then sculpt into something uniquely yours.

They’re especially useful for:

Brainstorming names when you’re stuck.

Testing how language feels in context.

Exploring cross-language ideas (e.g., “What would a French or Latin-inspired name sound like?”).

The secret isn’t to expect brilliance — it’s to use the LLM as a sounding board that helps you see what you like.

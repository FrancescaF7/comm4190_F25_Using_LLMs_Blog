I have always wondered if an LLM could be a good study buddy. Could it help me prepare for class, review readings, or even explain the topics that I cannot understand? 

I started by pasting in lecture notes from my political economy class last semester and asking: 

*"Can you quiz me on this material?"*





---

The model gave me a great selection of questions to test my knowledge. 

However, sometimes it would give me questions that did not come from my lectures notes. To fix this, I tried again: 

*"Only ask questions that directly come from the text I give you. Don't invesnt anything."*

That helped for the most part. Although, the questions became technically accurate but repetitive. 

* "What is the EU?"
* "Why does the EU matter?"
* "Explain the role of the EU"

This was not helpful and it felt as though I was being quizzed by a robot. 

I began to learn that the more direct inputs I gave it, the better exam questions it gave me. As such, once I told it to stop being repetitive, it gave me more great exam questions. 


Studying with an LLM feels a bit like talking to an over-eager person who is mostly helpful but occassionally makes things up. So you still have to be responsible by verifying, correcting, and deciding what to keep. 

The lesson here is that input really does matter, and can lead to a bad, good, or great outcome. In this way, when used right, an LLM can be an effective learning companion. It can help practice retrieval, organize chaos, and identify gaps.

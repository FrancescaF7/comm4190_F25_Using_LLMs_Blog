Budgeting apps are great in theory — until they start lecturing you about your “unusual coffee expenses.” So I wanted to see if Chat GPT 5 could act as a judgment-free version of a finance coach: one that just helps me figure out where my money goes and how to make it last.

---

I gave it a basic prompt:

“Here’s what I spent this week: $35 groceries, $28 coffee, $45 Uber, $60 eating out, $25 random Target run. Give me a summary and suggestions.”

It broke my spending down by category and politely pointed out that I had spent more on lattes than actual food. Then it reallocated my budget for the next week — not by cutting everything, but by suggesting small swaps (walk instead of Uber, cook once and reuse ingredients, skip one coffee run).

The tone felt realistic — not like a finance blog yelling about compound interest.

---

Next, I tried:

“Help me make a weekly budget as a student who spends around $150 per week. Prioritize food and transport.”

It created an easy table:

Groceries: $45

Transport: $30

Eating out: $25

Social/other: $20

Buffer: $30

Then it added suggestions like “batch cook lunches” and “check student discounts for Uber.” It wasn’t fancy, but it was practical — exactly what I’d want from a quick budgeting chat.

---

I took it one step further:

“At the end of each week, I’ll paste what I spent. Help me track patterns and adjust my categories.”

After pasting in two weeks worth, it started catching trends — like how I always overspent midweek when I was stressed. It even recommended setting a “fun budget” instead of pretending I wouldn’t spend on snacks or nights out.

It didn’t shame me — it adapted.

<p style="color:pink">LLMs can help you:</p>

* Categorize expenses clearly.
* Identify small habits that cost more than you think.
* Build personalized budgets that adjust over time.

The key is consistency. The more honest and detailed you are, the more useful the feedback becomes.

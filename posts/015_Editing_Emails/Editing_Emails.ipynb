Emails can be hard to write; sometimes they are too formal, too friendly or too try-hard. So I tested whether an LLM could help - not by writing from scratch, but by editing my drafts to sound more polite and confident. 

I gave ChatGPT 5 my original message:

“Hi Professor,
I was wondering if it’s okay to submit my reflection a few hours late — I had some issues with formatting. Sorry about that!!
Thank you,
Francesca”

Then I asked:

“Can you make this sound professional but still friendly?”

The model responded with a near-perfect rewrite:

“Dear Professor,
I hope you’re doing well. I wanted to ask if it would be possible to submit my reflection a few hours after the deadline due to a formatting issue. I really appreciate your understanding.
Best,
Francesca”

It was all correct but it had turned me into a polite robot.

---

Next, I prompted:

“Make this sound like me — a student who is respectful but warm and human.”

Now the tone shifted. The phrasing felt more natural, with just enough personality to sound real. The LLM had learned something subtle: not every message needs to sound like a corporate memo.

It made me realize that tone is where LLMs can actually help the most — not by automating communication, but by clarifying intention. I could experiment with tone sliders:

“More confident”

“More apologetic”

“More concise”

“Less passive”

Each tweak reflected a slightly different version of how I wanted to come across — almost like trying on different communication “outfits.”

---

In doing this exercise, I also discovered that LLMs can:

Summarize long threads so you don’t lose track of who said what.

Draft quick follow-ups (“Just checking in!” without the awkwardness).

Translate tone across audiences — turning “Hey!” into “Dear Hiring Manager.”

Proofread for tone shifts — catching when one sentence sounds friendly and the next sounds super serious. 

The key isn’t to let it replace your writing, but to treat it as a smart editor — one that never gets tired of rephrasing the same sentence.

<p style="color:purple">Reflection:</p>

Using an LLM for emails taught me that good writing isn’t about sounding perfect; it’s about aligning tone with intention. The model can help with grammar, structure, and even word choice — but it still can’t feel when “Best” sounds too stiff or when “Thanks!” feels too casual.

In a way, prompting an LLM to write like you is an exercise in self-awareness. You start noticing the small things — the phrases you overuse, the balance between warmth and professionalism, the emoji threshold for each relationship.

As such, the LLM won’t make you a better communicator automatically, but it will make you more conscious of how you already communicate.
